```python
SHADE = 3
WIDTH = 1600
HEIGHT = 1000


ROYAL_B0 = (170, 255, 72)
ROYAL_R0 = (248, 255, 222)
ROYAL_B1 = (150, 255, 84)
ROYAL_R1 = (252, 212, 238)

PALETTE_FR0 = [ROYAL_B0, (0, 0, 255), ROYAL_R0]
PALETTE_FR1 = [ROYAL_B1, (0, 0, 255), ROYAL_R1]


base0 = Image.new (mode="HSV", size=(WIDTH, HEIGHT), color=(0, 0, 100))

for i in range (SHADE):
    shadei = Image.new (mode="HSV", size=(int (WIDTH/SHADE), HEIGHT), color=PALETTE_FR0 [i])
    base0.paste (shadei, (i*int (WIDTH/SHADE), 0))
    

base1 = Image.new (mode="HSV", size=(WIDTH, HEIGHT), color=(0, 0, 100))

for i in range (SHADE):
    shadei = Image.new (mode="HSV", size=(int (WIDTH/SHADE), HEIGHT), color=PALETTE_FR1 [i])
    base1.paste (shadei, (i*int (WIDTH/SHADE), 0))
    
    
basecv0 = np.array (base0)
basecv1 = np.array (base1)
rc ["figure.figsize"] = (8, 4.5)

plt.subplot (121)
basecv2 = cv2.cvtColor (basecv0.copy (), cv2.COLOR_HSV2RGB_FULL)
plt.axis ("off")
plt.imshow (basecv2)

plt.subplot (122)
basecv3 = cv2.cvtColor (basecv1.copy (), cv2.COLOR_HSV2RGB_FULL)
plt.axis ("off")
plt.imshow (basecv3)
```

# Color palettes;

## Cyan, Magenta

```python
SHADE = 8
WIDTH = 1920
HEIGHT = 1080
HUE_CY = 128
HUE_MA = 234
V2, V1, V0 = 63, 127, 255 #BRIGHTNESS LEVELS
S1, S0 = 127, 255 # SATURATION LEVELS
PALETTE_CY = [(HUE_CY, S0, V2), (HUE_CY, S0, V1), (HUE_CY, S0, V0), (HUE_CY, S1, V0)]
PALETTE_MA = [(HUE_MA, S0, V2), (HUE_MA, S0, V1), (HUE_MA, S0, V0), (HUE_MA, S1, V0)]
PALETTE_0 = PALETTE_CY + PALETTE_MA


base_cyma = Image.new (mode="HSV", size=(WIDTH, HEIGHT), color=(0, 100, 100))

for i in range (SHADE):
    shadei = Image.new (mode="HSV", size=(int (WIDTH/SHADE), HEIGHT), color=PALETTE_0 [i])
    base_cyma.paste (shadei, (i*int (WIDTH/SHADE), 0))
    
    
basecv = np.array (base_cyma)
basecv_cyma = cv2.cvtColor (basecv.copy (), cv2.COLOR_HSV2RGB_FULL)

rc ["figure.figsize"] = (8, 4.5)
plt.axis ("off")
plt.imshow (basecv_cyma)
```

## Paste all palettes onto one 4K image

```python
WIDTH = 3840
HEIGHT = 2048

POS0 = (0, 0)
POS1 = (int (WIDTH/2), 0)
POS2 = (0, int (HEIGHT/2))
POS3 = (int (WIDTH/2), int (HEIGHT/2))

k4 = Image.new (mode="RGB", size=(WIDTH, HEIGHT), color=(0, 0, 100))
k4.paste (base_cyma, POS0)
k4.paste (base_bema, POS1)
k4.paste (base_blre0, POS2)
k4.paste (base_blre1, POS3)
                
k4.show ()
k4.save ("4 palettes original.png")
```
